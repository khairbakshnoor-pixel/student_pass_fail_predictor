```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("laptop.csv")
df
```

```python
from sklearn.preprocessing import StandardScaler
x=df.drop("Price",axis=1)
y=df['Price']
scale= StandardScaler()
x_scaled=scale.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
prediction=model.predict(x_test)
```

```python

# 1️⃣ Get input from the user
ram = int(input("Enter Size of ram: "))
rom = int(input("Enter size of rom: "))
age = int(input("Enter Age of the laptop: "))

# 2️⃣ Prepare input as a dataframe or array
import numpy as np
user_input = np.array([[ram,rom ,age]])


user_input_scaled = scale.transform(user_input)

# 4️⃣ Predict using the trained model
predicted_price = model.predict(user_input_scaled)

# 5️⃣ Show result
print(f"Predicted Price of the laptop: ${predicted_price[0]:,.2f}")
```
